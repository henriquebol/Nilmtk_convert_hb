```python
import json
import pandas as pd
from pandas.io.json import json_normalize as jn

pd.set_option('display.float_format', lambda x: '%.5f' % x)


import numpy as np

csv = '/var/snap/node-red/675/teste/pzem_arduino_2hz_4dias.csv'
geral = pd.read_csv(csv, sep=',')
geral = geral.drop(columns=['Millis'])
geral['Time'] = pd.to_datetime(geral['Time'],unit='ms')
geral = geral.set_index('Time').resample("1s").mean()
# print (geral.describe())

csv = '/var/snap/node-red/675/teste/microondas.csv'
micro = pd.read_csv(csv, sep=',')
# micro = micro.drop(columns=['ReactivePower','Factor','Voltage','Current'])
micro = micro.rename(columns={'Power': 'Active(micro)', 
                         'ApparentPower': 'ApparentPower(micro)', 
                         'ReactivePower': 'ReactivePower(micro)', 
                         'Factor': 'Factor(micro)', 
                         'Voltage': 'Voltage(micro)', 
                         'Current': 'Current(micro)'})
micro['Time'] = pd.to_datetime(micro['Time'],unit='ms')
micro = micro.set_index('Time')
# print (micro.describe())

csv = '/var/snap/node-red/675/teste/ventilador.csv'
vent = pd.read_csv(csv, sep=',')
# vent = vent.drop(columns=['ReactivePower','Factor','Voltage','Current'])
vent = vent.rename(columns={'Power': 'Active(vent)', 
                         'ApparentPower': 'ApparentPower(vent)', 
                         'ReactivePower': 'ReactivePower(vent)', 
                         'Factor': 'Factor(vent)', 
                         'Voltage': 'Voltage(vent)', 
                         'Current': 'Current(vent)'})
vent['Time'] = pd.to_datetime(vent['Time'],unit='ms')
vent = vent.set_index('Time')

geral.describe()
```

### Concatenando (micro e vent)

```python
concatenado_micro = pd.concat([geral,micro], axis=1)
concatenado_micro.dropna(subset = ["Active(micro)"], inplace=True) # Removendo NaN
concatenado_micro.dropna(subset = ["Power"], inplace=True) # Removendo NaN
concatenado_micro.loc['2021-09-02 11:55:41':'2021-09-02 11:55:50']

concatenado_vent = pd.concat([geral,vent], axis=1)
concatenado_vent.dropna(subset = ["Active(vent)"], inplace=True) # Removendo NaN
concatenado_vent.dropna(subset = ["Power"], inplace=True) # Removendo NaN
# concatenado_vent.loc['2021-09-02 11:55:41':'2021-09-02 11:55:50']
```

```python
concatenado_micro['label'] = np.where((concatenado_micro['Active(micro)']>40) , 1, 0)
concatenado_vent['label'] = np.where((concatenado_vent['Active(vent)']>40) , 1, 0)
# concatenado_micro
```

```python
# display 
import seaborn as sb
sb.heatmap(concatenado_micro.astype('float64').corr(), annot=True, cmap="Blues")
sb.pairplot(concatenado_micro, hue='label', diag_kind='kde') # Densidades
```
